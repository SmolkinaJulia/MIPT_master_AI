## Часть 3: жадный алгоритм раскраски

Найдите для заданного `networkx`-графа $G$ раскраску не более, чем в $\Delta(G)$ цветов с помощью рассказанного на лекции «жадного» алгоритма. Технически раскраска — это словарь, в котором для каждой вершины указан её номер цвета от $1$ до $\Delta(G)$.

```python
def greedy_coloring(G):
    color_map = {}
    # Consider nodes in descending degree 
    for node in sorted(G, key=lambda x: len(G[x]), reverse=True):
        neighbor_colors = set(color_map.get(neigh) for neigh in G[node])
        color_map[node] = next( 
            color for color in range(len(G)) if color not in neighbor_colors
        )
    return color_map
```

```python
# Простой тест алгоритма раскраски (запустите перед отправкой):
#import sys
#import traceback
G = nx.random_regular_graph(7, 10, 2020)
coloring = greedy_coloring(G)
nx.draw_circular(G,with_labels=True,edge_color='b',node_color = list(coloring))
assert max(coloring.values()) <= 7
    
assert min(coloring.values()) >= 0
     
assert all(coloring[u] != coloring[v] for u,v in G.edges(G))
```
